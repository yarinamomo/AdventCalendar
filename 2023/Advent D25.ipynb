```python
com_input = {}
with open("data/input25.txt","r") as file:
    for line in file.readlines():
        m1, ms = line.strip().split(':')
        com_input[m1] = tuple([i for i in ms.strip().split(' ')])
```

```python
import networkx as nx

G = nx.Graph(com_input)
```

```python
res = sorted(map(sorted, nx.k_edge_components(G, k=4)))
```

```python
len(res)
```

```python
len(res[0])*len(res[1])
```
