#### Part 1

```python
import numpy as np

def read_file():
    input_file_path = "data/input17.txt"
    registers = {"A":0, "B":0, "C":0}
    program = []
    def process_line(line, if_upper):
        nonlocal program
        if if_upper:
            id_re = line.strip().split(":")[0][-1]
            registers[id_re] = int(line.strip().split(":")[1])
        else:
            program = [int(i) for i in line.strip().split(":")[1].split(",")]

    with open(input_file_path, 'r') as file:
        if_upper = True
        for line in file:
            if not line.strip():
                if_upper = False
                continue
            if if_upper:
                process_line(line, True)
            else:
                process_line(line, False)
    return registers, program
```

```python
def combo_operand(operand_in):
    if 0 <= operand_in <= 3:
        return operand_in
    if operand_in == 4:
        return registers["A"]
    if operand_in == 5:
        return registers["B"]
    if operand_in == 6:
        return registers["C"]
    if operand_in == 7:
        print("invalid operand 7")
        return None
    print("invalid operand", operand_in)
    return None

def adv(operand_in):
    operand = combo_operand(operand_in)
    registers["A"] = int(registers["A"]/pow(2, operand))
def bxl(operand_in):
    registers["B"] = registers["B"]^operand_in
def bst(operand_in):
    operand = combo_operand(operand_in)
    registers["B"] = operand % 8
def jnz(operand_in):
    if registers["A"] != 0:
#         print("pointer jumps to", operand_in)
        return operand_in
def bxc(operand_in):
    registers["B"] = registers["B"]^registers["C"]
def out(operand_in):
    operand = combo_operand(operand_in)
    outputs.append(operand % 8)
#     print("outputing", outputs)
def bdv(operand_in):
    operand = combo_operand(operand_in)
    registers["B"] = int(registers["A"]/pow(2, operand))
def cdv(operand_in):
    operand = combo_operand(operand_in)
    registers["C"] = int(registers["A"]/pow(2, operand))

instructions = {0: adv, 1: bxl, 2: bst, 3: jnz, 4: bxc, 5: out, 6: bdv, 7: cdv}
def instruct(opcode, operand_in):
    global ins_pointer, registers
    argument = {"operand_in": operand_in}
    pointer = instructions[opcode](**argument)
    if pointer != None:
        ins_pointer = pointer
    else:
        ins_pointer += 2
```

```python
ins_pointer = 0
outputs = []
registers, program = read_file()

while (ins_pointer < len(program)-1):
    opcode = program[ins_pointer]
    operand_in = program[ins_pointer + 1]
#     print(ins_pointer, opcode, operand_in, registers)
    instruct(opcode, operand_in)

",".join(map(str,outputs))
```
